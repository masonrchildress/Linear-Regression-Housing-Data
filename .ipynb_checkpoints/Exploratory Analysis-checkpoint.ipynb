```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import r2_score

%matplotlib inline

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('./datasets/cleaned_train.csv')
no_dum_df = pd.read_csv('./datasets/no_dum_cleaned_train')
```

```python
# Several features with top correlation with SalePrice
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(df.corr()[['SalePrice']].sort_values('SalePrice').tail(20),
            vmax=1, vmin=-1, cmap='YlGnBu', annot=True, ax=ax);
ax.invert_yaxis()
plt.savefig('./images/best10_cmap.png')
```

```python
# Several features with bottom correlation with Gv Liv Area
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(df.corr()[['Gr Liv Area']].sort_values('Gr Liv Area').tail(20),
            vmax=1, vmin=-1, cmap='YlGnBu', annot=True, ax=ax);
ax.invert_yaxis()
plt.title()
plt.savefig('./images/worst10_cmap.png')
```

```python
# Scatter of Gr Liv Area and SalePrice sorted by Fireplace Qu
g= sns.scatterplot('Gr Liv Area', 'SalePrice', data=no_dum_df)
plt.title('Above Ground Living Area vs Sale Price')
plt.xlabel('Above Ground Living Area (Square Feet)')
plt.ylabel('Sale Price')
g.figure.savefig('liv_area_v_price')
```

```python
no_dum_df.groupby('Fireplace Qu').mean().sort_values('SalePrice')[['SalePrice']].plot(kind='barh')
plt.ylabel('Fireplace Quality')
plt.xlabel('Average Sale Price')
plt.title('Quality of Fireplace and Average Sale Price');
ax = plt.gca()
labels = [item.get_text() for item in ax.get_yticklabels()]
ratings = ['Poor','None','Fair','Average','Good','Excellent']
for i in range(0,6):
    labels[i] = ratings[i]
ax.set_yticklabels(labels)
plt.savefig('./images/fireplace_qu_v_price');
```

```python
no_dum_df.groupby('Fireplaces').mean().sort_values('SalePrice')[['SalePrice']].plot(kind='barh')
plt.title('Number of Fireplaces and Average Sale Price')
plt.ylabel('Fireplace Quantity')
plt.xlabel('Sale Price');
plt.savefig('./images/num_fireplace_v_price');
```
